```python
import pandas as pd
data={
    "name":['rohan','suhil','jade'],
    'brach':['cs','cp','at'],
    'math':[10,40,60],
     'science':[87,40,66]
}
df=pd.DataFrame(data)
df['total']=df[['math','science']].sum(axis=1)
df['average']=df[['math','science']].mean(axis=1)
df['result']=df['average'].apply(lambda x:'pass' if x>=40 else'fail')
df.to_csv('student result.csv',index='fals')
print(df)
```
